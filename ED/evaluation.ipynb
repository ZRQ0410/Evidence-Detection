### Evaluation

We use dev.csv file to evaluate the performance for the svm model and bi-gru model.

Files requried for evaluation:
- dev.csv file
- output file (prediction) of the svm mode
- output file (prediction) of the bi-gru model

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import classification_report
```

Read the development file:

```python
dev = pd.read_csv('dev.csv')
labels = dev['label']
labels.head()
```

Read the output (prediction) file:

```python
# svm model output
pred_svm = pd.read_csv('pred_svm.csv')
# bi-gru model output
pred_gru = pd.read_csv('pred_gru.csv')
```

```python
def model_eval(labels, pred, round=3):
    acc = float(np.round(accuracy_score(labels, pred), 3))
    # calculate metrics for each label, and find their unweighted mean
    precision = float(np.round(precision_score(labels, pred, average='macro'), 3))
    recall = float(np.round(recall_score(labels, pred, average='macro'), 3))
    f1 = float(np.round(f1_score(labels, pred, average='macro'), 3))
    auc = float(np.round(roc_auc_score(labels, pred), 3))
    rpt = classification_report(labels, pred)

    print('Accuracy:', acc)
    print('Precision:', precision)
    print('Recall:', recall)
    print('F1:', f1)
    print('ROC AUC:', auc)
    print('Classification report:')
    print(rpt)
```

```python
# evaluation for svm
print('Prediction evaluation of svm model:\n')
model_eval(labels, pred_svm)

print('-' * 60)
print('Prediction evaluation of bi-gru model:\n')
# evaluation for bi-gru
model_eval(labels, pred_gru)
```
